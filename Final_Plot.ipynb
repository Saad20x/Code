```python
import numpy as np
import pandas as pd
import folium as fo
df = pd.read_csv('C:/Users/Saad/Downloads/Alassandro Discord/gdansk_august.csv')
df.head()
```

```python
lat = list(df["latitude"])
lon = list(df["longitude"])
```

```python
volcano = fo.FeatureGroup(df="Volcano")
for a, b, in zip(lat, lon):
  volcano.add_child(fo.Marker(location=[a, b], icon=fo.Icon(color='blue')))
  
fo.Map().add_child(volcano)
```

```python
import pandas as pd
import numpy as np
import seaborn as sb

df = pd.read_csv('C:/Users/Saad/Downloads/Alassandro Discord/poland_august.csv')
df
```

```python
import folium
```

```python
lats = df['latitude']
lons = df['longitude']
```

```python
a = folium.Map(location=[lats[0], lons[0]])
```

```python
for lat, lon in zip (lats, lons):
    
    folium.Marker(location=[lat, lon],
    popup = lat, lon,
    icon=folium.Icon(color='green')).add_to(a)
```
